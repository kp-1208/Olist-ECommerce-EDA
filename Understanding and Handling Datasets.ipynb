```python
# Merge length,height and width into one attribute named size.
products['size'] = products['product_length_cm'] * products['product_height_cm'] * products['product_width_cm']
products.drop(['product_length_cm','product_height_cm','product_width_cm'], axis =1, inplace = True)
```

```python
# Handling null values
products = products.drop(['product_name_lenght','product_description_lenght','product_photos_qty'], axis =1)
products.product_weight_g.fillna(method = 'ffill', inplace = True)
products.product_length_cm.fillna(method = 'ffill', inplace = True)
products.product_height_cm.fillna(method = 'ffill', inplace = True)
products.product_width_cm.fillna(method = 'ffill', inplace = True)
```
