### Q2. Customer who has purchased products of maximum amount.

```python
order_amounts = payments.groupby(['order_id'], as_index=False).sum()
order_amounts = (order_amounts.merge(orders, left_on='order_id', right_on='order_id')
                 .reindex(columns=['customer_id', 'payment_value']))
order_amounts = (order_amounts.merge(customers, left_on='customer_id', right_on='customer_id')
                 .reindex(columns=['customer_unique_id', 'payment_value']))
order_amounts = order_amounts.groupby(['customer_unique_id'],as_index=False).sum()
order_amounts.sort_values(by=['payment_value'], ascending= False,inplace=True) 
order_amounts.head(10) # Top 10 customers in terms of revenue.
```

### Q3. How much share he has in company's revenue?

```python
order_amounts['revenue%'] = (order_amounts['payment_value']/payments.payment_value.sum())*100
order_amounts.head(10)
```

### Q4. Is he/she a regular customer?

```python
cust_orders = order_amounts.head(10)
cust_orders = (cust_orders.merge(customers, left_on='customer_unique_id', right_on='customer_unique_id')
                 .reindex(columns=['customer_id','customer_unique_id', 'payment_value']))
```

```python
# A new customer_id is generated for each order a customer makes. So the number of occurences of customer_unique_id 
# indicates the number of orders a customer has made.
cust_orders_freq = cust_orders.groupby(['customer_unique_id'],as_index=False).count()
cust_orders_freq = cust_orders_freq[['customer_unique_id','customer_id']]
cust_orders_freq.rename(columns={'customer_id':'no_of_orders'},inplace=True)
cust_orders_freq = (cust_orders_freq.merge(order_amounts.head(10), left_on='customer_unique_id', right_on='customer_unique_id')
                 .reindex(columns=['customer_unique_id', 'payment_value','revenue%','no_of_orders']))
cust_orders_freq.sort_values(by=['payment_value'], ascending= False,inplace=True) 
cust_orders_freq
```

### Q5. What review score does he/she give on the orders?

```python
c_review = (cust_orders.merge(orders, left_on='customer_id', right_on='customer_id')
                 .reindex(columns=['customer_id', 'order_id']))
c_review = (c_review.merge(reviews, left_on='order_id', right_on='order_id')
                 .reindex(columns=['customer_id', 'order_id', 'review_score']))
c_review = (c_review.merge(cust_orders, left_on='customer_id', right_on='customer_id')
                 .reindex(columns=['customer_unique_id', 'customer_id', 'review_score']))
c_review = c_review.groupby(['customer_unique_id'],as_index=False).mean()
c_review = (c_review.merge(cust_orders_freq, left_on='customer_unique_id', right_on='customer_unique_id')
                 .reindex(columns=['customer_unique_id', 'payment_value','revenue%','no_of_orders','review_score']))
c_review.sort_values(by=['payment_value'], ascending= False,inplace=True) 
c_review
```

### MOST VALUABLE CUSTOMER BASED ON ANALYSIS OF ABOVE PARAMETERS.

```python
c_value = c_review
c_value['customer_value_score'] = c_review['payment_value']*c_review['revenue%']*c_review['no_of_orders']*c_review['review_score']
c_value.sort_values(by=['customer_value_score'], ascending= False,inplace=True)
c_value
```

### Q7. a) Maximum revenue from which area?

```python
data = (customers.merge(orders, left_on='customer_id', right_on='customer_id')
                 .reindex(columns=['order_id','customer_city']))
data = (data.merge(payments, left_on='order_id', right_on='order_id')
                 .reindex(columns=['order_id','customer_city','payment_value']))
top_revenue_cities = data.groupby("customer_city",as_index=False)["payment_value"].sum().sort_values("payment_value",
                                                                                                     ascending = False)
plt.figure(figsize = (14, 7))
sns.barplot(x = "payment_value", y = "customer_city", data = top_revenue_cities[:10])
plt.xlabel("Total Revenue (in Millions)")
plt.ylabel("Cities")
plt.title("Top 10 cities with highest revenue.")
plt.show()
top_revenue_cities = top_revenue_cities[:10]
top_revenue_cities.rename(columns={'payment_value':'total_revenue'},inplace=True)
top_revenue_cities
```

### Q7. b) Minimum revenue from which area?

```python
data = (customers.merge(orders, left_on='customer_id', right_on='customer_id')
                 .reindex(columns=['order_id','customer_city']))
data = (data.merge(payments, left_on='order_id', right_on='order_id')
                 .reindex(columns=['order_id','customer_city','payment_value']))
top_revenue_cities = data.groupby("customer_city",as_index=False)["payment_value"].sum().sort_values("payment_value")
plt.figure(figsize = (14, 7))
sns.barplot(x = "payment_value", y = "customer_city", data = top_revenue_cities[:20])
plt.xlabel("Total Revenue (in Millions)")
plt.ylabel("Cities")
plt.title("Bottom 20 cities with least revenue.")
plt.show()
top_revenue_cities = top_revenue_cities[top_revenue_cities.payment_value<1000]
top_revenue_cities.rename(columns={'payment_value':'total_revenue'},inplace=True)
top_revenue_cities
```

### Q8. Which is the most sold product category in this area?

```python
temp = (top_orders_cities.merge(customers, left_on='customer_city', right_on='customer_city')
                 .reindex(columns=['customer_id', 'customer_city']))
temp = (temp.merge(orders, left_on='customer_id', right_on='customer_id')
                 .reindex(columns=['customer_id', 'order_id','customer_city']))
temp = (temp.merge(order_items, left_on='order_id', right_on='order_id')
                 .reindex(columns=['customer_id', 'order_id','product_id','customer_city']))
temp = (temp.merge(products, left_on='product_id', right_on='product_id')
                 .reindex(columns=['order_id', 'product_id','product_category_name','customer_city']))
temp = temp.groupby(['customer_city'],as_index=False)['product_category_name'].agg(pd.Series.mode)
temp = (temp.merge(top_orders_cities, left_on='customer_city', right_on='customer_city')
                 .reindex(columns=['customer_city','no_of_orders','product_category_name']))
temp.rename(columns={'product_category_name':'most_sold_category'},inplace=True)
temp.sort_values("no_of_orders", ascending = False,inplace=True)
temp
```

### Q9. Average delivery time in this area?

```python
carrier_possession = orders[['customer_id','order_approved_date','order_delivered_customer_date']]
carrier_possession = (carrier_possession.merge(customers, left_on='customer_id', right_on='customer_id')
                 .reindex(columns=['customer_id','customer_city','order_approved_date','order_delivered_customer_date']))
carrier_possession['average_delivery_time'] = carrier_possession.order_delivered_customer_date - carrier_possession.order_approved_date
carrier_possession = carrier_possession.groupby(['customer_city'],as_index=False)['average_delivery_time'].mean()
carrier_possession = (temp.merge(carrier_possession, left_on='customer_city', right_on='customer_city')
                 .reindex(columns=['customer_city','no_of_orders','most_sold_category','average_delivery_time']))
carrier_possession
```

### Q12. Whose fault in late delivery, seller or logistic partner?

```python
late = orders[orders.order_delivered_customer_date > orders.order_estimated_delivery_date]
late['fault'] = 0
late.fault[late.order_delivered_carrier_date > late.order_estimated_delivery_date] = 'seller'
late.fault[late.order_delivered_carrier_date <= late.order_estimated_delivery_date] = 'logistic partner'
late.drop(late.loc[late['fault']==0].index, inplace=True)
plt.figure(figsize = (14, 7))
sns.countplot(x='fault', data = late)
plt.ylabel("No of late delivered orders")
plt.show()
```
