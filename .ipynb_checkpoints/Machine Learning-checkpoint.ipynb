# Applying Machine Learning for classification and prediction using the dataset.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

customers = pd.read_csv("final_dataset/customers.csv")
order_items = pd.read_csv("final_dataset/order_items.csv")
orders = pd.read_csv("final_dataset/orders.csv")
payments = pd.read_csv("final_dataset/payments.csv")
products = pd.read_csv("final_dataset/products.csv")
reviews = pd.read_csv("final_dataset/reviews.csv")
sellers = pd.read_csv("final_dataset/sellers.csv")
```

```python
# Converting object to DateTime format.

orders['order_purchase_date'] =  pd.to_datetime(orders['order_purchase_date'], format="%d-%m-%Y %H:%M")
orders['order_approved_date'] =  pd.to_datetime(orders['order_approved_date'], format="%d-%m-%Y %H:%M")
orders['order_delivered_carrier_date'] =  pd.to_datetime(orders['order_delivered_carrier_date'], format="%d-%m-%Y %H:%M")
orders['order_delivered_customer_date'] =  pd.to_datetime(orders['order_delivered_customer_date'], format="%d-%m-%Y %H:%M")
orders['order_estimated_delivery_date'] =  pd.to_datetime(orders['order_estimated_delivery_date'], format="%d-%m-%Y %H:%M")
```

### Predicting the fault of logistic partner in late delivery of order.

```python
late = orders[orders.order_delivered_customer_date > orders.order_estimated_delivery_date]
late['fault'] = 0
late.fault[late.order_delivered_carrier_date > late.order_estimated_delivery_date] = 'seller'
late.fault[late.order_delivered_carrier_date <= late.order_estimated_delivery_date] = 'logistic partner'
late.drop(late.loc[late['fault']==0].index, inplace=True)
late
```

```python
late = late[['order_purchase_date','order_delivered_carrier_date','order_delivered_customer_date',
             'order_estimated_delivery_date','fault']]
late
```

```python
# Convert the datetime columns to integer 
```

```python
late['order_purchase_date'] = late['order_purchase_date'].apply(lambda x:x.toordinal())
late['order_delivered_carrier_date'] = late['order_delivered_carrier_date'].apply(lambda x:x.toordinal())
late['order_delivered_customer_date'] = late['order_delivered_customer_date'].apply(lambda x:x.toordinal())
late['order_estimated_delivery_date'] = late['order_estimated_delivery_date'].apply(lambda x:x.toordinal())
late
```

```python
late['purchase_to_carrier'] = late['order_delivered_carrier_date'] - late['order_purchase_date']
late['carrier_to_customer'] = late['order_delivered_customer_date'] - late['order_delivered_carrier_date']
late = late[['purchase_to_carrier','carrier_to_customer','fault']]
late.fault[late.fault == 'seller'] = 0
late.fault[late.fault == 'logistic partner'] = 1
late
```

```python
# Loading different binary classification models.

models = {}

# Logistic Regression
from sklearn.linear_model import LogisticRegression
models['Logistic Regression'] = LogisticRegression()

# Support Vector Machines
from sklearn.svm import LinearSVC
models['Support Vector Machines'] = LinearSVC()

# Decision Trees
from sklearn.tree import DecisionTreeClassifier
models['Decision Trees'] = DecisionTreeClassifier()

# Random Forest
from sklearn.ensemble import RandomForestClassifier
models['Random Forest'] = RandomForestClassifier()

# Naive Bayes
from sklearn.naive_bayes import GaussianNB
models['Naive Bayes'] = GaussianNB()

# K-Nearest Neighbors
from sklearn.neighbors import KNeighborsClassifier
models['K-Nearest Neighbor'] = KNeighborsClassifier()
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn.metrics import accuracy_score, precision_score, recall_score


#split dataset in features and target variable
feature_cols = ['purchase_to_carrier','carrier_to_customer']
X = late[feature_cols] # Features
y = late['fault'] # Target Variables
y=y.astype('int')

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 70% training and 30% test

accuracy, precision, recall = {}, {}, {}

for key in models.keys():
    
    # Fit the classifier
    models[key].fit(X_train, y_train)
    
    # Make predictions
    predictions = models[key].predict(X_test)
    
    # Calculate metrics
    accuracy[key] = accuracy_score(predictions, y_test)
    precision[key] = precision_score(predictions, y_test)
    recall[key] = recall_score(predictions, y_test)
    
results = pd.DataFrame(index=models.keys(), columns=['Accuracy', 'Precision', 'Recall'])
results['Accuracy'] = accuracy.values()
results['Precision'] = precision.values()
results['Recall'] = recall.values()

results
```
